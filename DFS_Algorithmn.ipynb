```python
def dfs(graph, node, visited=None):
    if visited is None:
        visited = set()
    
    visited.add(node)
    print(node, end=" ")  # Print the current node
    
    for neighbor in graph.get(node, []):
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Define the graph as an adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G'],
    'E': [],
    'F': [],
    'G': []
}

# Perform DFS traversal starting from node A
print("Depth-First Search Traversal:")
dfs(graph, 'A')
```

```python
from collections import deque

def generate_neighbors(word, word_bank):
    neighbors = []
    for i in range(len(word)):
        for char in 'abcdefghijklmnopqrstuvwxyz':
            new_word = word[:i] + char + word[i+1:]
            if new_word != word and new_word in word_bank:
                neighbors.append(new_word)
    return neighbors

def shortest_transformation(start_word, end_word, word_bank):
    word_bank.add(end_word)
    queue = deque([(start_word, [start_word])])
    visited = set([start_word])
    
    while queue:
        current_word, path = queue.popleft()
        if current_word == end_word:
            return path
        
        neighbors = generate_neighbors(current_word, word_bank)
        for neighbor in neighbors:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    
    return None

# Example usage
start_word = "hit"
end_word = "cog"
word_bank = {"hot", "dot", "dog", "lot", "log"}

result = shortest_transformation(start_word, end_word, word_bank)
if result:
    print("Shortest Transformation Sequence:")
    print(" -> ".join(result))
else:
    print("No transformation sequence exists.")
```
